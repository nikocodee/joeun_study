```python
# 구구단 출력
for i in range(2, 10):  # 2단부터 9단까지
    for j in range(1, 10):  # 1부터 9까지 곱하기
        print(f"{i} x {j} = {i * j}")
    print("-" * 20)  # 구분선
```

```python
for i in range(1,4): # 1~3
    for j in range(1,3): # 1~2
        if (i > j):
            print("*", end="")
    print("*")  
```

```python
for i in range(1,4): # 1~3
    for j in range(1,3): # 1~2
        if (i > j):
            print("", end="")
        else:
            print("*", end="")
    print("*")  
```

```python
for i in range(1,4): # 1~3
    for j in range(1,3): # 1~2
        if (i > j):
            print("*".rjust(2), end="")
    print("*".rjust(2)) 
```

```python
lines = [
    "This is a test",
    "Python is fun",
    "I love python programming"
]

keyword = "Python"
filtered_lines = ['/'+ line + '|' for line in lines if keyword.lower() in line.lower()]
print(filtered_lines)
```

```python
# 문자열 -> bytes
text = "Hello"
bytes_data = text.encode("utf-8")
print(bytes_data)

# bytes -> 문자열
decoded_text = bytes_data.decode("utf-8")
print(decoded_text)
```

```python
nums = []
for i in range(100):
    try:
        div = i - 50
        #print(i, div)
        nums = 100/div
    # 정확한 error명 알 때    
    except ZeroDivisionError as ze: 
        print(f"Error i : {i} {ze}")
    # 정확한 error명 모를때    
    except Exception as ex:
        print("Error!!!!")    
print("End")
```

```python
nums = []
for i in range(100):
    try:
        div = i - 50
        #print(i, div)
        nums = 100/div
    # 정확한 error명 알 때    
    except IndexError as ie: 
        print(f"Error i : {i} {ie}")
    # 정확한 error명 모를때    
    except Exception as ex:
        print("Error!!!!")    
print("End")
```
